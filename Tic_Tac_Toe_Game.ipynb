```python
import numpy as np
matrix = np.array([["-", "-", "-"],
                 ["-", "-", "-"],
                 ["-", "-", "-"]])

def win():
    if(matrix[0][0] == matrix[0][1] == matrix[0][2] == "X" or matrix[0][0] == matrix[1][0] == matrix[2][0] == "X"  or matrix[0][0] == matrix[1][1] == matrix[2][2] == "X"
      or matrix[1][0] == matrix[1][1] == matrix[1][2] == "X" or matrix[2][0] == matrix[2][1] == matrix[2][2] == "X"  or matrix[0][1] == matrix[1][1] == matrix[2][1] == "X"
      or matrix[0][2] == matrix[1][2] == matrix[2][2] == "X" or matrix[2][0] == matrix[1][1] == matrix[0][2] == "X"):
        print("player 1 wins!!!!!!")
        return True
        
    elif(matrix[0][0] == matrix[0][1] == matrix[0][2] == "O" or matrix[0][0] == matrix[1][0] == matrix[2][0] == "O"  or matrix[0][0] == matrix[1][1] == matrix[2][2] == "O"
      or matrix[1][0] == matrix[1][1] == matrix[1][2] == "O" or matrix[2][0] == matrix[2][1] == matrix[2][2] == "O"  or matrix[0][1] == matrix[1][1] == matrix[2][1] == "O"
      or matrix[0][2] == matrix[1][2] == matrix[2][2] == "O" or matrix[2][0] == matrix[1][1] == matrix[0][2] == "O"):
        
        print("player 2 wins!!!!!!")
        return True
        
    else:
        return False
        
def place(symbol):
    print(matrix)
    entry = True
    while entry:
        try:
            row = int(input("Enter the row"))
            col = int(input("Enter the column"))
            if row > -1 and row < 3 and col > -1 and col < 3 and matrix[row][col] == "-":
                break
            else:
                print("Enter a valid entry")
                entry = True
        except:
            print("Enter valid Integer value only")
            entry = True
    matrix[row][col] = symbol
            
        
        
    
        

p1 = "X"
p2 = "O"
print("-----------TIC TAC TOE GAME --------------")
def play():
    for turn in range(9):
        if turn%2 == 0:
            print("Player X turn")
            place(p1)
            if win():
                break
        else:
            print("Player O turn")
            place(p2)
            if win():
                break
    if not win():
        print("Draw")
play()
```
